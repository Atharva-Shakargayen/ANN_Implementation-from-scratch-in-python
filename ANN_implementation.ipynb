```python
dfy=pd.DataFrame(y_transposed)
dfy
```

```python
# Define a mapping function to convert binary pair to label
def map_binary_to_label(row):
    if row[0] == 0 and row[1] == 1:
        return 'Normal'
    elif row[0] == 1 and row[1] == 0:
        return 'Cancer'
    else:
        return 'Unknown'  # In case there are any other combinations

# Apply the function to each row and create a new column
dfy['N/C'] = dfy.apply(map_binary_to_label, axis=1)

print("\nDataFrame with Target Labels:")
print(dfy[['N/C']])
```

```python
y=dfy[['N/C']]
y
```

```python
y.value_counts() # label is not skewed
```

```python
y = pd.get_dummies(y, drop_first=True)
y.sample(5) # cancer --> 1 and normal --> 0
```

```python
X = df.drop('N/C_Normal', axis=1)
y = df['N/C_Normal']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
X_train.head()
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=100, activation='relu'),
     keras.layers.Dense(1, activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test, y_pred))
```

```python
### AUC and ROC Curve, confusion matrix and precision recall curve
auc = tf.keras.metrics.AUC()
auc.update_state(y_test, y_pred)
print("AUC: ", auc.result().numpy())
```

```python
roc=tf.keras.metrics.AUC(curve='ROC')
roc.update_state(y_test, y_pred)
print("ROC: ", roc.result().numpy())
```

```python
confusion_matrix(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(y_test, y_pred)
import matplotlib.pyplot as plt
plt.plot(recall, precision)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```
